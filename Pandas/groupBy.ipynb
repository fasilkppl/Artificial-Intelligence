```python
#this function returns the key if condition is met
def grouper_func(df, idx, colname):
    if 80<= df[colname].loc[idx] <= 90:
        return '80 to 90'
    elif 50<= df[colname].loc[idx] <= 60:
        return '50 to 60'
    else:
        return 'others'
```

```python
#i wanna group all rows that have temp between 80 to 90, 50 to 60 then print
df_obj = df.groupby(lambda idx: grouper_func(df, idx, 'temperature'))
```

```python
#the groupby df obj always will have key and data when we use it in loop
for key, data in df_obj:
    print("Key : ", key)
    print("\n")
    print('DF :', data)
    print('\n')
    
```
