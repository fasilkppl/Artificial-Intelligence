```python
import pandas as pd
df = pd.read_csv('climate_data.csv')
df
```

```python
#replace -99999 with 0 (old value, new value)
df.replace(-99999,0)
```

```python
#replace -99999 with NaN
import numpy as np
df.replace(-99999,np.nan)
```

```python
#replace -99999 and -88888 with NaN
df.replace([-99999, -88888], np.nan)
```

```python
#if we want we can specific keyword arguments
df.replace(to_replace = [-99999, -88888], value = np.nan)
```

```python
#we can also specify the columns to replace values and specific what to replace with eg. this case np.nan
df.replace({
    'temperature' : -99999,
    'windspeed' : [-99999, -88888],
    'event' : 'no event'
}, np.nan)
```

```python
#in all columns whenever we find -99999, -88888, no event replace them with given value
df.replace({
    -99999 : np.nan,
    -88888 : np.nan,
    'no event' : 'sunny day'
})
```
