```python
#add new column based on 1 existing column(in this case release_year)
#age = current year - release_year
#lambda is like a loop variable on every iteration it will store the release year of that sepcific movie

df["age"] = df.release_year.apply(lambda x: 2025 - x) 
df.head(3)
```

```python
#OR
df["age"] = df["release_year"].apply(lambda x: 2025 - x) 
df.head(3)
```

```python
#OR
df["age"] = df.apply(lambda x: 2025 - x["release_year"], axis=1) 
df.head(2)
```

```python
#add new column based on 2 existing columns (in this case revenue and budget)
#profit = revenue - budget
df["profit"] = df.apply(lambda x: x["revenue"] - x["budget"], axis = 1)
df.head(2)
```
