```python
import pandas as pd
df = pd.read_csv("weather_data.csv", parse_dates = ['day'])
type(df.day[0]) #covert to pandas date series
```

```python
df.set_index('day', inplace = True)
df
```

```python
df.fillna(0, inplace=True)
df #inplace = True modifies original dataframe
```

```python
#modifies n.a values with mean() values or median() values
df = pd.read_csv("weather_data.csv")
df.fillna({
    'temperature' : df.temperature.mean(),
    'windspeed' : df.temperature.median(),
    'event' : 'No Event'
})
```

```python
#fills the gives current value to next NaN value
df = pd.read_csv("weather_data.csv")
df.fillna(method='ffill')
```

```python
df = pd.read_csv("weather_data.csv")
df.fillna(method='bfill', axis=1) # colums values given to backwards
```

```python
#fills the gives current value to next NaN value(only 1 next value)
df = pd.read_csv("weather_data.csv")
df.fillna(method='ffill', limit = 1)
```

```python
#equally increments the values
df.interpolate()
```

```python
df.dropna() #drops all rows that have  n.a in them
```

```python
#specifies the minimum number of non-NA values required to keep a row or column.
df.dropna(thresh=2)
```
