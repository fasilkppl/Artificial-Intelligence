```python
import pandas as pd
us_wheather = pd.DataFrame({
    'city': ['new york', 'chicago', 'orlando'],
    'temperature': [23, 45, 38],
    'humidity': [34,78,56]
})
```

```python
uae_wheather = pd.DataFrame({
    'city': ['dubai', 'Abu dhabi', 'Sharjah'],
    'temperature': [34,44, 55],
    'humidity': [67,77,99]
})
```

```python
#concatinate two df, vertical mount
#ignore_index create squential index other wise it will be 0,1,2,0,1,2
pd.concat([us_wheather, uae_wheather], ignore_index=True) 
```

```python
#we can give key and later use them to get all data under that key
df = pd.concat([us_wheather, uae_wheather], keys=['usa', 'uae']) 
df
```

```python
#we can give key and later use them to get all data under that key
df.loc['uae']
```

```python
cars_1 = pd.DataFrame({
    'carName': ['bugatti', 'lambo', 'Lexus'],
    'colour': ['red','green', 'yellow'],
}, index=[0,1,2]) #we use index to make the index same of both tables

cars_1
```

```python
pd.concat([cars_1, cars_2], axis = 1) #horizontal mount
```

```python
df_1 = pd.DataFrame({
    'city': ['dubai', 'Abu dhabi', 'Sharjah'],
    'humidity': [67,77,99]
})
```

```python
df_2 = pd.DataFrame({
    'city': ['dubai', 'Abu dhabi'],
    'temperature': [34,44],
})
```

```python
#inner join - only gives both common fields (cities)
pd.merge(df_1, df_2, on='city')
```

```python
pd.merge(df_1, df_2, on='city')
```

```python
#left join - gives all fields on left then the common fields from right
pd.merge(df_1, df_2, on='city', how='left')
```

```python
#right join - gives all fields on right then the common fields from left
pd.merge(df_1, df_2, on='city', how='right')
```

```python
#outer join - gives all fields on right and left
pd.merge(df_1, df_2, on='city', how='outer')
```
