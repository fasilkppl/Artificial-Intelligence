```python
'''This is a dataset that describes sonar chirp returns bouncing off different services. 
The 60 input variables are the strength of the returns at different angles. 
It is a binary classification problem that requires a model to differentiate rocks from metal cylinders.'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("./sonar_dataset.csv", header=None)
df.sample(5)

# this d fdoesnt have column names, so, when i loaded the df, i used header = None
```

```python
df.shape
```

```python
# check for nan values
df.isna().sum()

# results is 0, so no null values
```

```python
df.columns
```

```python
# i have 111 M samples and 97 R samples and 60 columns
df[60].value_counts() # label is not skewed
```

```python
#Splitting as X and Y, droping column 60
X = df.drop(60, axis='columns')
y = df[60]
y.head()
```

```python
# our y have M and R letters, so One hot encoding
y = pd.get_dummies(y, drop_first=True)
y.sample(5) # R --> 1 and M --> 0
```

```python
y.value_counts()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
X_train.head()
```

```python
# Using Deep Learning Model
```

```python
# Model without Dropout Layer
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

Training Accuracy >>> Test Accuracy

```python

y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
# comparing y_predicted and truth (y_test)
# as you can see some mistakes are there
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```

```python
# Model with Dropout Layer
# we add drop layer between hidden layer
```

```python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

modeld.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

modeld.fit(X_train, y_train, epochs=100, batch_size=8) # batch_size=8 → model learns from 8 samples at a time
```

```python
'''
Creating the model container
“Sequential” means layers are stacked one after another | Data flows in one direction: input → output
First Dense layer (input + hidden layer)
keras.layers.Dense(60, input_dim=60, activation='relu'),
Dense → fully connected layer
60 → number of neurons in this hidden layer | input_dim=60 → each input sample has 60 features | activation='relu' → applies the ReLU function
This is: The input layer Also the first hidden layer

3. Dropout layer (after first Dense layer)
keras.layers.Dropout(0.5) | Randomly disables 50% of neurons during training | Helps prevent overfitting | Only active during training, not during prediction

4. Second Dense layer (hidden layer)
keras.layers.Dense(30, activation='relu') | Fully connected layer | 30 neurons in this hidden layer | Uses ReLU activation | No input_dim needed because Keras already knows the shape

5. Dropout layer (after second Dense layer)
keras.layers.Dropout(0.5) | Again disables 50% of neurons | Adds regularization to this layer

6. repeats this

7. Output layer
keras.layers.Dense(1, activation='sigmoid') | 1 neuron (output layer) → because this is binary classification | sigmoid outputs a value between 0 and 1 | Interpreted as a probability
'''
```

```python
# above is the accuracy using train data.
# accuracy using test data.
modeld.evaluate(X_test, y_test)
```

```python
#Training Accuracy is still good but Test Accuracy Improved
```

```python
y_pred = modeld.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```

```python
# You can see that by using dropout layer test accuracy increased from 0.77 to 0.81
```
