```python
class HashTable:
    def __init__ (self):
        self.MAX = 100
        self.arr = [None for i in range(100)]
    def get_hash(self, key):
        h=0
        for char in key:
            h+=ord(char)
        return h % self.MAX
    def add_item_into_array_space(self, key, value):
        h = self.get_hash(key) # get hash from key
        self.arr[h] = value #assign it in array space using hash as array(index).
    def get_value_from_key(self,key):
        h = self.get_hash(key)
        return self.arr[h]
```

```python
class HashTable:
    def __init__ (self):
        self.MAX = 100
        self.arr = [None for i in range(10)]
    def get_hash(self, key):
        h=0
        for char in key:
            h+=ord(char)
        return h % self.MAX
    def __setitem__(self, key, value):
        h = self.get_hash(key) # get hash from key
        self.arr[h] = value #assign it in array space using hash as array(index).
    def __getitem__ (self,key):
        h = self.get_hash(key)
        return self.arr[h]
    def __delitem__ (self, key):
        h = self.get_hash(key)
        self.arr[h] = None
```

```python
obj2 = HashTable()
obj2['bob'] = 250 #set item -these are magic methods __ double underscore so python calls them behind the scenes but we need to spcify some things using objects like eg. obj2['bob']
obj2['bobb'] = 350 #set item
obj2['bobbb'] = 450 #set item
obj2['bobbbb'] = 550 #set item
```
