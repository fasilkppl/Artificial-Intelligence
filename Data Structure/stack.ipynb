```python
class Stack: 
    def __init__(self):
        self.container = deque()
    def push(self, value):#adds on top
        self.container.append(value)
    def pop(self): #removes last value permanently
        return self.container.pop()
    def peek(self): # returns last value, doesnt remove it.
        return self.container[-1]
    def is_empty(self):
        return len(self.container) == 0
    def size(self):
        return len(self.container)
    def show(self):
        return list(self.container)
        
```

```python
s = Stack()
s.push("happy wife")
s.push("happy life")
s.push("Naah..")
s.push("Just Kidding")
```

```python
s.pop()
```

```python
s.peek()
```

```python
s.show()
```
