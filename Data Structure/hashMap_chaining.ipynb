```python
class HashTable:
    def __init__ (self):
        self.MAX = 10
        self.arr = [[] for i in range(10)]
    def get_hash(self, key):
        h=0
        for char in key:
            h+=ord(char)
        return h % self.MAX
        
    def __setitem__(self, key, value):
        h = self.get_hash(key) # get hash from key

        found = False
        for i, element in enumerate(self.arr[h]):
            if element[0] == key:
                self.arr[h][i] = (key,value)
                found = True
                break
        if not found:
            self.arr[h].append((key, value))
                
            
    def __getitem__ (self,key):
        h = self.get_hash(key)
        for element in self.arr[h]:
            if element[0] == key:
                return element[1] 
        return None
        
    def __delitem__ (self, key):
        h = self.get_hash(key)
        for i,element in enumerate(self.arr[h]):
            if element[0] == key:
                del self.arr[h][i]
                break
```

```python
obj = HashTable()
obj.get_hash("march 6")
```

```python
obj.get_hash("march 17")
```
