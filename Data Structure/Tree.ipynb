```python
class TreeNode:
    def __init__(self, data):
        self.data = data # the name of the item
        self.parent = None # parent node
        self.children = [] # list of children
        
    def add_child(self, child): # child is HP, Dell, Samsung etc..
        child.parent = self # tells the child who its parent is.
        self.children.append(child) #adds the child to the parentâ€™s list. "Laptop" now has "HP" in its children list
    
    def get_level(self): #This counts how deep the node is in the tree.
        level = 0
        p = self.parent
        while p:
            level+=1
            p = p.parent #p = p.parent moves p up the tree to the next parent.
        return level
        
    def print_tree(self):
        spaces = " " * self.get_level() * 3
        prefix = spaces + "___|"

        if self.parent:
            prefix = spaces + "___|"
            print(prefix + self.data)
        else:
            print(self.data)

        if self.children:
            for child in self.children:
                child.print_tree()
```

```python
def build_product_tree():
    root = TreeNode("Electronics")

    laptop = TreeNode("Laptop")
    laptop.add_child(TreeNode("HP"))
    laptop.add_child(TreeNode("Dell"))
    laptop.add_child(TreeNode("Asus ROG"))
    
    cellphone = TreeNode("Cell Phone")
    cellphone.add_child(TreeNode("Realme"))
    cellphone.add_child(TreeNode("Apple"))
    cellphone.add_child(TreeNode("Xioami"))

    tv = TreeNode("Television")
    tv.add_child(TreeNode("LG"))
    tv.add_child(TreeNode("Sony"))
    tv.add_child(TreeNode("Samsung"))

    root.add_child(laptop)
    root.add_child(cellphone)
    root.add_child(tv)

    #calling print function
    root.print_tree()
```

```python
if __name__ == '__main__':
    build_product_tree()
```
