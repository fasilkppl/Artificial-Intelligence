```python
#Gradient Descent to find the best straight line (linear regression) that fits the points:

# x: [1,2,3,4,5]
# y: [5,7,9,11,13]
# m (slope) and b (intercept).
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 150
    learning_rate = 0.01
    n = len(x)

    plt.scatter(x, y, color='red', marker='+', linewidth=5) #plot points (red colour points)

    for i in range(iterations):
         # ðŸ‘‰ plot BEFORE updating
        y_predicted = m_curr * x + b_curr

        cost = (1/n) * sum((y - y_predicted)**2)
        print("m_curr:",m_curr,"|b_curr:",b_curr,"|cost:",cost)
        

        m_partial = -(2/n) * sum(x * (y - y_predicted))
        b_partial = -(2/n) * sum(y - y_predicted)

        m_curr -= learning_rate * m_partial
        b_curr -= learning_rate * b_partial

        # ðŸ‘‰ plot each line as the model updates
        plt.plot(x, y_predicted, color='green', linewidth=1) #plot lines,(green lines) 

        # optional print
        # print(i, m_curr, b_curr, cost)

    plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
#function call
gradient_descent(x,y)
```
