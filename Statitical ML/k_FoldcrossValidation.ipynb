```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
digits = load_digits()
```

```python
digits.data
```

```python
digits.target
```

```python
# You donâ€™t need to drop any columns when using load_digits()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target,test_size=0.3)
```

```python
# Logistic Regression
lr = LogisticRegression(solver='liblinear',multi_class='ovr')
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
# SVM
svm = SVC(gamma='auto')
svm.fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
# Random Forest
rf = RandomForestClassifier(n_estimators=40)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# Doing KFold cross validation
from sklearn.model_selection import KFold
kf = KFold(n_splits=3)
kf
```

```python
# this splits the data set given [1,2,3,4,5,6,7,8,9] to 3 equal parts 
# uses 1 fold for testing and others for training
# on second iteration it changes the testing and training data, till k no. of numbers.
for train_index, test_index in kf.split([1,2,3,4,5,6,7,8,9]):
    print(train_index, test_index)
```

```python
# creted a method called get score.
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
# like this, we can get the score of any model.
get_score(LogisticRegression(), X_train, X_test, y_train, y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=3)
```

```python
# This code is the internal working code, not required while doing actual k fold cross validation
# creating three lists
scores_logistic = []
scores_svm = []
scores_rf = []

for train_index, test_index in folds.split(digits.data,digits.target):
    X_train, X_test, y_train, y_test = digits.data[train_index], digits.data[test_index], \
                                       digits.target[train_index], digits.target[test_index]
    scores_logistic.append(get_score(LogisticRegression(solver='liblinear',multi_class='ovr'), X_train, X_test, y_train, y_test))  
    scores_svm.append(get_score(SVC(gamma='auto'), X_train, X_test, y_train, y_test))
    scores_rf.append(get_score(RandomForestClassifier(n_estimators=40), X_train, X_test, y_train, y_test))
```

```python
scores_rf
```

```python
# cross_val_score function in real life doing
from sklearn.model_selection import cross_val_score
```

```python
#Logistic regression model performance using cross_val_score
# cv is the no. of folds in cross validation, here 3 fold cross validation
cross_val_score(LogisticRegression(solver='liblinear',multi_class='ovr'), digits.data, digits.target,cv=3)
```

```python
#svm model performance using cross_val_score
cross_val_score(SVC(gamma='auto'), digits.data, digits.target,cv=3)
```

```python
# random forest performance using cross_val_score
cross_val_score(RandomForestClassifier(n_estimators=40),digits.data, digits.target,cv=3)
```

```python
# Parameter tunning using k fold cross validation
scores1 = cross_val_score(RandomForestClassifier(n_estimators=5),digits.data, digits.target, cv=10)
np.average(scores1)
```
