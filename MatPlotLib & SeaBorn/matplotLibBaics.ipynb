```python
import pandas as pd
df_sales = pd.read_excel('linechart.xlsx')
df_sales.head(3)
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12,7))
plt.title("Product Sales")
#(x axis, y axis)
plt.plot(df_sales['Quarter'], df_sales['Fridge'], color = 'blue', label= 'Fridge')
plt.plot(df_sales['Quarter'], df_sales['Dishwasher'], color = 'green', label= 'Dishwasher')
plt.plot(df_sales['Quarter'], df_sales['Washing Machine'], color = 'red', label= 'Washing Machine')

plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)

plt.show()
```

```python
total_sales = df_sales[["Fridge", "Dishwasher", "Washing Machine"]].sum()
```

```python
total_sales.index
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct = '%1.1f%%', explode = (0.1,0,0), shadow = True, startangle= 140)
plt.show()
```
